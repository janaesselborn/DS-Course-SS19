```python
import pandas as pd
import os
import json
files = os.listdir("data/return-data")
print(files)
```

```python
df = pd.DataFrame()

allData = "["
for filename in files:
    filepath = "data/return-data/" + filename
        
    with open(filepath, "r") as fp:
        for line in fp:
            data = json.loads(line)
            
            # basket,returnLabel,totalAmount,transactionId,zipCode
            dataformA = {'basket': data['basket'], 'returnLabel': data['returnLabel'], 'totalAmount': data['totalAmount'], 'transactionId': data['transactionId'], 'zipCode': data['zipCode']}
            allData = allData + str(dataformA) + ","
            
allData += "]"
df = pd.DataFrame(eval(allData))        
print(df)
```

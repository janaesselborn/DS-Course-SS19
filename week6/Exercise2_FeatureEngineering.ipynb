```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
# Use function
# Is there an easier way, then that?
df['c_0'] = df['basket'].apply(lambda x: x.count("0"))
df['c_1'] = df['basket'].apply(lambda x: x.count("1"))
df['c_2'] = df['basket'].apply(lambda x: x.count("2"))
df['c_3'] = df['basket'].apply(lambda x: x.count("3"))
df['c_4'] = df['basket'].apply(lambda x: x.count("4"))
df['c_5'] = df['basket'].apply(lambda x: x.count("5"))
df.head()
```

```python
# Drop the original basket
train = df.drop(columns=['basket'])
train.head()
```

```python
# Preprocessing the data
x = train.drop(columns=["returnLabel"])
scaler = StandardScaler()
X_std = scaler.fit_transform(x)
y = train['returnLabel'].values

# Create a testsplit
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.30,random_state=101)
```

```python
# Do the logistic Regression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)

from sklearn.metrics import classification_report, accuracy_score
print(classification_report(y_test,predictions))
print("Accuracy:", accuracy_score(y_test, predictions))
```

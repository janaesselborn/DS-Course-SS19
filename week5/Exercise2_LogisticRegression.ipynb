```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv('data.csv')
train = train.drop(columns=['basket'])
train.head()
```

```python
train_one_hot = pd.get_dummies(train["zipCode"])
train_one_hot.head()
```

```python
# Concat train and train_one_hot
train_new = pd.concat([train.drop(columns=["zipCode"]), train_one_hot], axis=1)
train_new.head()
```

```python
from sklearn.model_selection import train_test_split

y = train_new['returnLabel'].values
X_train, X_test, y_train, y_test = train_test_split(train_new, y, test_size=0.30,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)

from sklearn.metrics import classification_report, accuracy_score
print(classification_report(y_test,predictions))
print("Accuracy:", accuracy_score(y_test, predictions))
```
